```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
```

```python
df = load_iris()

X = pd.DataFrame(df.data)
y = df.target
```

```python
X.head()
```

```python
scaler = StandardScaler()
X_scale = scaler.fit_transform(X)
X_scale.shape
# The number features here are 4
```

# PCA

```python
from sklearn.decomposition import PCA

pca = PCA(
    n_components=2
)
X_pca = pca.fit_transform(X_scale)
X_pca.shape
# Now the number of features are 2.
```

```python
# variance 
print(f"Explained Variance Ratio for PC-1 and PC-2: ", pca.explained_variance_ratio_)
```

```python
# eigen vectors.
print(f"For PC-1 and PC-2: \n" , pca.components_)
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    X_pca[:,0],
    X_pca[:,1],
)

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA for iris dataset")
# Now this pc1 will become X axis and pc2 will become Y axis by rorating this axis
```
