```python
from sklearn.datasets import load_iris
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
df = load_iris()

X = df.data
y = df.target
```

```python
scaler = StandardScaler()
X_scale = scaler.fit_transform(X)
```

```python
# Visualize
sns.scatterplot(x=X_scale[:,0], y=X_scale[:,1])
```

```python
# DB SCAN
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(
    eps=0.8,
    min_samples=5
)
```

```python
labels = dbscan.fit_predict(X_scale)
sns.scatterplot(x=X_scale[:,0], y=X_scale[:,1], c=labels)
# Here in the below plot 4 datapoints colored in purple are noise/outliers
```

# DB scan on NON LINEAR DATA

```python
from sklearn.datasets import make_moons
```

```python
X, y = make_moons(
    n_samples=300,
    noise=0.05,
    random_state=42
)
```

```python
scaler = StandardScaler()
X_scale = scaler.fit_transform(X)
```

```python
sns.scatterplot(x=X_scale[:,0], y=X_scale[:,1])
```

```python
dbscan = DBSCAN(
    eps=0.5,
    min_samples=5
)
labels = dbscan.fit_predict(X_scale)
sns.scatterplot(x=X_scale[:,0], y=X_scale[:,1], c=labels)
```

# Another example of DB SCAN on non linear data

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_circles
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
X, y = make_circles(
    n_samples=300,
    noise=0.05,
    factor=0.4,
    random_state=42
)
```

```python
scaler = StandardScaler()
X_scale = scaler.fit_transform(X)

plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
dbscan = DBSCAN(
    eps=0.5,
    min_samples=5
)
labels = dbscan.fit_predict(X_scale)
plt.scatter(X_scale[:, 0], X_scale[:, 1], c=labels)
plt.show()
```
