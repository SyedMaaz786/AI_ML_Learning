```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv("heart.csv")
df.head()
```

```python
X = df.drop(columns = "target", axis=1)
y = df["target"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

scaler = StandardScaler()
X_train_scale = scaler.fit_transform(X_train)
X_test_scale = scaler.transform(X_test)
X_train_scale
```

```python
# k=3
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train_scale, y_train)
```

```python
y_pred = model.predict(X_test_scale)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
print("accuracy:", accuracy)
print("precision:", precision)
print("recall:", recall)
```

```python
# k=5
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train_scale, y_train)
```

```python
y_pred = model.predict(X_test_scale)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
print("accuracy:", accuracy)
print("precision:", precision)
print("recall:", recall)
```

```python
# k=7
model = KNeighborsClassifier(n_neighbors=7)
model.fit(X_train_scale, y_train)
```

```python
y_pred = model.predict(X_test_scale)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
print("accuracy:", accuracy)
print("precision:", precision)
print("recall:", recall)
```

```python
# Cross validation for hyperparameter tuning using GridSearchCV.

from sklearn.model_selection import GridSearchCV

model = KNeighborsClassifier()
param_grid = {"n_neighbors": [3,5,7,9]}

model_cv = GridSearchCV (
    model,
    param_grid,
    cv=5,
    scoring="recall"
)

model_cv.fit(X_train_scale, y_train)

y_pred = model_cv.predict(X_test_scale)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
print("accuracy:", accuracy)
print("precision:", precision)
print("recall:", recall)

#results
res = pd.DataFrame(model_cv.cv_results_)
print(res)
print(model_cv.best_params_) 
```

```python
# Pipeline 
# Remember for pipeline we dont have to pass scale data like x_train_scale, just pass xtrain and ytrain pipeline will automatically do the scaling 

from sklearn.pipeline import Pipeline

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

pipe = Pipeline([("scaler", StandardScaler()), ("knn", KNeighborsClassifier())])

param_grid = {"knn__n_neighbors": [3,5,7,9]}  #Here knn__ means we are just appending this to our neighbors

model_cv = GridSearchCV (
    pipe,
    param_grid,
    cv=5,
    scoring="recall"
)

model_cv.fit(X_train, y_train)

y_pred = model_cv.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
print("accuracy:", accuracy)
print("precision:", precision)
print("recall:", recall)

#results
res = pd.DataFrame(model_cv.cv_results_)
print(res)
print(model_cv.best_params_) #remember this is based on recall
```
