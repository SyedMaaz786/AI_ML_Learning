```python
import gymnasium as gym
import numpy as np
import random
```

```python
gamma = 0.99
epsilon = 0.1
alpha = 0.5
episodes = 500
Q = np.zeros((48, 4)) # states , actions
```

```python
# Epsilon-policy
def epsilon_greedy(state):
    if random.random() < epsilon:
        return env.action_space.sample() # EXPLORE
    else:
        return np.argmax(Q[state]) # EXPLOIT
```

```python
# Q-learning

for episode in range(episodes):

    env = gym.make("CliffWalking-v1")
    done = False
    state, _ = env.reset()

    total_reward = 0
    episode_len = 0

    while not done:
        action = epsilon_greedy(state)
        next_state, reward, terminated, truncated, _ = env.step(action)
        done = terminated or truncated

        # Q-learning update
        # This is the formula check notes
        Q[state, action] += alpha * (reward + gamma * np.max(Q[next_state]) - Q[state, action])

        state = next_state

        total_reward += reward
        episode_len += 1

    print(f"episode = {episode+1} / {episodes}, total reward = {total_reward} & episode length = {episode_len}")
    env.close()
    
```
