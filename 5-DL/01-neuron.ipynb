```python
import torch
import torch.nn as nn
```

```python
# inputs (creating i/p features).

torch.manual_seed(42) # for same o/p everytime like random_state in ML
inputs = torch.tensor([1.0, 2.0, 3.0])
```

```python
neuron = nn.Linear(in_features=3, out_features=1)
```

```python
output = neuron(inputs)
```

```python
output
```

```python
print(neuron.weight)
print(neuron.bias)
```

```python
# verify ... (formula)

inputs @ neuron.weight.T + neuron.bias
```
