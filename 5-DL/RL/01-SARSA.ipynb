# SARSA

```python
gamma = 0.99
alpha = 0.5
epsilon = 0.1
episodes = 500
```

```python
# Q talks => stores Q-values
Q = np.zeros((48, 4)) # states , actions
Q
```

```python
# Policy - Epsilon-greedy : pass State as i/p and it returns action as o/p

def epsilon_greedy(state):
    if random.random() < epsilon:
        return env.action_space.sample() # random action => EXPLORE
    else:
        return np.argmax(Q[state]) # EXPLOIT
```

```python
for episode in range(episodes):  # remember we have not written the code of rendering the training process which will unnecessary take time and not imp but if you want refer lecture or GPT
    
    env = gym.make("CliffWalking-v1")
    done = False # When the episode has ended
    state, _ = env.reset()
    action = epsilon_greedy(state)

    total_reward = 0
    episode_len = 0 # moves in particular episode

    while not done:
        next_state, reward, terminated, truncated, _ = env.step(action) # With this action being passed in the env it returns these values => next_state, reward, terminated, truncated  
        done = terminated or truncated
        next_action = epsilon_greedy(next_state)

        # SARSA update
        # This is the formula check notes
        Q[state, action] += alpha * (reward + gamma * (Q[next_state, next_action] - Q[state, action]))  
        
        state = next_state
        action = next_action

        total_reward += reward
        episode_len += 1
    print(f"Episode = {episode + 1} / {episodes}, total reward = {total_reward} & episode length = {episode_len}")
    env.close()
```

```python
# What did our agent learn?

env = gym.make("CliffWalking-v1", render_mode = "human")
state, _ = env.reset()
done = False
total_reward = 0
episode_len = 0

while not done:
    action = np.argmax(Q[state])
    next_state, reward, terminated, truncate, _ = env.step(action)
    done = terminated or truncate

    state = next_state

    total_reward += reward
    episode_len += 1

print(f"total reward = {total_reward} & episode length = {episode_len}")
env.close()
```
