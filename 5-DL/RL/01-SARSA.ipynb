```python
import gymnasium as gym # Used to create env objects
import numpy as np
import random
```

```python
# Methods in gymnasium include ...
# make() - for creating env
# render() - for showing env
# reset() - for reseting env
# close() - for closing env
# step() - for taking action in env
# There are many more refer doc ...
```

```python
# creating env
env = gym.make("CliffWalking-v1")
env
```

```python
print(f"States: {env.observation_space.n}") # states
print(f"Actions: {env.action_space.n}") # actions

starting_state, _ = env.reset() # We need row,col val which gives 2 vals, but we need only 1 val as state, so we use this formula (row * 12 * col)  
print(f"Starting state: {starting_state}")
```
