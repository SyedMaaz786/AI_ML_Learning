```python
#Practice problem

# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]
years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

plt.plot(years, oscar_revenue)

#Matplotlib methods
plt.title("Oscar Movies Revenue In Each Year")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
```

```python
# Non-Oscar Winner Movies
non_oscar_movies = [
    "Slumdog Millionaire",
    "Avatar",
    "Inception",
    "Hugo",
    "Lincoln"
]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

plt.plot(years, oscar_revenue, color = "red", marker = "<", linestyle = "--", linewidth = 3, label="Oscar Movies")
plt.plot(years, non_oscar_revenue, "<--g", label="Non-Oscar Movies")

#Matplotlib methods
plt.title("Oscar Movies VS Non-Oscar Movies Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
plt.legend()

plt.style.available
plt.style.use("dark-background")
plt.style.use("ggplot")
plt.style.use("default")
plt.show()
plt.grid()
plt.tight_layout()
plt.savefig("name_of_plot.png")
```

```python
# Bar Plot (Vertical)

oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]
years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

plt.bar(years, oscar_revenue)
plt.title("Revenue In Each Year For Oscar Movies")
plt.xlabel("Year")
plt.ylabel("Renevue (in $M)")

#Bar chart methods

plt.text(years[0], oscar_revenue[0], str(oscar_revenue[0]), ha="center")

for i in range(len(years)):
    plt.text(years[i], oscar_revenue[i]+20, str(oscar_revenue[i]), ha="center")

plt.ylim(0, max(oscar_revenue)+100)
```

```python
import numpy as np

oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

non_oscar_movies = [
    "Slumdog Millionaire",
    "Avatar",
    "Inception",
    "Hugo",
    "Lincoln"
]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M
years = [2008, 2009, 2010, 2011, 2012]

x = np.arange(len(years))
width = 0.3

plt.title("Oscar VS Non-Oscar Movies Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
plt.bar(x-width/2, oscar_revenue, width, label="Oscar Revenue") #Sub means shifting to LEFT
plt.bar(x+width/2, non_oscar_revenue, width, label="Non-Oscar Revenue") #Add means shifting ot RIGHT
plt.legend()

plt.xticks(x, years)  #Here we have converted the year(len) values back to the real label values
```

```python
# Scatter plot

people = [
    "Person A", "Person B", "Person C", "Person D", "Person E",
    "Person F", "Person G", "Person H", "Person I", "Person J"
]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
bp = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]

colors = ["green" if x < 135 else "red" for x in bp]
plt.scatter(age, bp, s=bp, cmap="OrRd", c=bp, marker="<") #Here cmap is the plasma
plt.title("Age VS BP")
plt.xlabel("Age")
plt.ylabel("BP")
plt.grid()
plt.colorbar(label="BP")

for i in range(len(people)):
    plt.annotate(people[i], xy=(age[i], bp[i]), xytext=(age[i]+1, bp[i]+1))
plt.xlim(min(age), max(age)+10)
plt.ylim(min(bp), max(bp)+5)

#I did this for practise
plt.text(age[0], bp[0], f"Age: ({age[0]}, BP: {bp[0]})")
```

```python
cities = ["City A", "City B", "City C", "City D", "City E"]

# Winter: Temperature (C) vs Humidity (%)
winter_temp = [5, 2, 10, 0, 7]
winter_humidity = [80, 75, 65, 85, 70]

# Summer: Temperature (C) vs Humidity (%)
summer_temp = [25, 30, 28, 35, 27]
summer_humidity = [60, 50, 55, 45, 65]

plt.scatter(winter_temp, winter_humidity, label="Winter")
plt.scatter(summer_temp, summer_humidity, label="Summer")
plt.title("Winter VS Summer Data")
plt.xlabel("Temp")
plt.ylabel("Humidity")
plt.legend()
```

```python
# Pie chart

expenses = ["Salaries", "Rent", "Marketing", "R&D", "Miscellaneous"]
amounts = [500, 150, 120, 100, 50]

explode = [0,0,0,0.2,0]

plt.pie(amounts, labels=expenses, autopct="%1.1f%%",
       wedgeprops = {
           "edgecolor": "black",
           "linewidth": 2,
           "linestyle": "--"
       },
        explode = explode,
        startangle = 90,
        shadow = True
       ) 
plt.show()
```

```python
# Histogram

import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)

scores = np.random.normal(70, 10, 100) #normal distribution
plt.hist(scores, color="skyblue", edgecolor="black", bins=5)
plt.axvline(
    x=70,
    label="Average marks (70)",
    color="red",
    linewidth="2",
    linestyle="--"
)
plt.xlabel("Scores")
plt.ylabel("Students")
```

```python
# multiple datasets in histogram
legit_transactions = [
    2.99, 5.49, 8.99, 12.50, 14.99, 19.99, 23.45, 29.99, 34.99, 39.50,
    45.00, 49.99, 55.25, 60.00, 75.99, 89.99, 120.50, 150.00, 199.99,
    249.99, 300.75, 450.00, 600.00, 850.00, 1200.00
]

fraud_transactions = [
    50, 100, 150, 200, 300, 500, 500, 750, 1000, 1000,
    1200, 1500, 1500, 2000, 2500, 3000, 3000
]

#Transaction amount distribution - Legit vs Fraud

plt.hist(legit_transactions, color="green", bins=10, label="Legit", edgecolor="black", alpha=0.5)
plt.hist(fraud_transactions, color="red", bins=10, label="Fraud", edgecolor="black", alpha=0.5)

plt.title("Transaction Amount Distribution - Legit vs Fraud")
plt.xlabel("Amount in $")
plt.ylabel("Frequency")
plt.legend()
```
