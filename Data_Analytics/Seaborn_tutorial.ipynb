```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")  #Here we have loaded tips to plot the data for the same for learning
print(type(tips))
print(tips.head())
```

```python
#plotting data using seaborn

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    style="smoker"
)
```

```python
#Scatter plot in seaborn

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="time"
)
```

```python
#Line plot in seaborn

flights = sns.load_dataset("flights")
sns.lineplot(
    data=flights,
    x="year",
    y="passengers"
)
```
