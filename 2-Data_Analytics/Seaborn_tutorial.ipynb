```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")  #Here we have loaded tips to plot the data for the same, for learning
print(type(tips))
print(tips.head())
```

```python
#plotting data using seaborn

sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    style="smoker"
)
```

```python
#Scatter plot in seaborn

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="time"
)
```

```python
#Line plot in seaborn

flights = sns.load_dataset("flights")
sns.lineplot(
    data=flights,
    x="year",
    y="passengers"
)
```

```python
#Bar plot in seaborn

sns.barplot(
    data=tips,
    x="day",
    y="tip",
    hue="sex"
)
```

```python
#Box plot in seaborn

sns.boxplot(
    data=tips,
    x="day",
    y="tip"
)
```

```python
# histogram plot in seaborn

penguins = sns.load_dataset("penguins")

sns.histplot(
    data=penguins,
    x="body_mass_g",
    bins=30
)
```

```python
#Heat map 

flight_pivot = flights.pivot(index="month", columns="year", values="passengers") # Here we have converted the long format to short (check lecture once)

sns.heatmap(
    flight_pivot,
    cmap="coolwarm",
    annot=True,
    fmt="d"
)

plt.title("Passengers Data")
plt.tight_layout()
```
